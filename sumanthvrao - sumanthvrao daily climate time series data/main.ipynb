```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import vispy as vp
```

```python
data = pd.read_csv('Data/DailyDelhiClimateTrain.csv')
data_units = {
    "meantemp": "Â°C",
    "humidity": "g/m3",
    "wind_speed": "km/h",
    "meanpressure": "mbar",
}

data.index = pd.to_datetime(data['date'], format='%Y-%m-%d')
del data["date"]
n = len(data)
```

```python
for col in data.columns:
    data[col].plot()
    plt.title(col + " - " + data_units[col])
    plt.show()

for col in data.columns:
    data[col].hist(bins=50)
    plt.title(col + " - " + data_units[col])
    plt.show()
```
